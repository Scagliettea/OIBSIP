```python
#Rename columns. 
#I found that the column names are incorrect after looking over the missing values. Thus, I'm going to rename every column to help you comprehend this data better.

data2.columns= ['States','Date','Frequency',
               'Estimated Unemployment Rate',
               'Estimated Employed',
               'Estimated Labour Participation Rate',
               'Region','longitude','latitude']
```

```python
#Checking the correlation between the feature of dataset.

heatmaps = data2[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'longitude', 'latitude']]
heatmaps = heatmaps.corr()
plt.figure(figsize=(10,6))
sns.set_context('notebook', font_scale=1)
sns.heatmap(data=heatmaps, annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
#Analysis of Unemployment Rates: Data Visualization
#In order to examine the unemployment rate, let's now plot the data. I'll start by looking at the approximate number of workers by India's various regions:

data2.columns= ['States','Date','Frequency',
               'Estimated Unemployment Rate','Estimated Employed',
               'Estimated Labour Participation Rate','Region',
               'longitude','latitude']
plt.figure(figsize=(10, 6))
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed', hue='Region', data=data2)
plt.show()
```

```python
#Let's now examine the unemployment rate by India's various regions.

plt.figure(figsize=(10, 6))
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Unemployment Rate', hue='Region', data=data2)
plt.show()
```

```python
#Unemployment rate of each Indian state by region dashboard.
#Let's now construct a dashboard to examine the unemployment rate by region in each Indian state. I'll utilize a sunburst layout for this.

unemployment = data2[['States', 'Region', 'Estimated Unemployment Rate']]
figure = px.sunburst(unemployment, path=["Region", "States"], 
                     values='Estimated Unemployment Rate', 
                     height=550, color_continuous_scale='Rdylbu', 
                     title='Unemployment Rate in India')
figure.show()
```
